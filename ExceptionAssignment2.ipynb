Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
 
The ArithmeticError class in Python is a base class for various arithmetic-related exception classes. It serves as a superclass for exceptions that occur during arithmetic operations. Some of the notable errors defined in the ArithmeticError class include:

ZeroDivisionError: Raised when attempting to divide a number by zero.
OverflowError: Raised when an arithmetic operation exceeds the limits of the data type.

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error:", e)
```

```python
import math

try:
    result = math.exp(1000)
except OverflowError as e:
    print("Error:", e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

 The LookupError class in Python is a base class for exceptions that occur when an invalid key or index is used to access a container, such as a dictionary or a list. It is a superclass for various lookup-related exceptions, including KeyError and IndexError. Using LookupError as a base class provides a common way to handle these specific lookup-related exceptions.

```python
person = {"name": "John", "age": 30, "city": "New York"}

try:
    print(person["gender"])
except KeyError as e:
    print("Key Error:", e)
```

```python
numbers = [1, 2, 3, 4, 5]

try:
    print(numbers[10])
except IndexError as e:
    print("Error:", e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is a built-in Python exception that occurs when there is an issue while importing a module or a specific attribute from a module.


ModuleNotFoundError is a subclass of ImportError.
It is more specific than ImportError and is raised when the interpreter cannot find the module being imported.

```python
try:
    import non_existent_module
except ImportError as e:
    print("Import Error:", e)
```

```python
try:
    import non_existent_module
except ModuleNotFoundError as e:
    print("Module Not Found Error:", e)
```
